```python
from diffusers import StableDiffusionPipeline
import torch

pipe = StableDiffusionPipeline.from_pretrained(
    "runwayml/stable-diffusion-v1-5"
)

pipe = pipe.to("cpu")

prompts = [
    "A futuristic AI-powered classroom with holographic displays, ultra realistic, 4k",
    "Advanced AI healthcare center with robotic doctors, 3D render",
    "Smart farming with AI drones monitoring crops, cartoon style",
    "Green sustainable city powered by renewable energy, minimalist design",
    "Cyberpunk AI city with intelligent robots, neon lights"
]

for i, prompt in enumerate(prompts):
    image = pipe(prompt).images[0]
    image.save(f"poster{i+1}.png")

print("All posters generated!")
```

```python
from gtts import gTTS

text = """
Welcome to the Generative AI Internship Program.
Today we explore AI-powered image and voice generation technologies.
"""

tts = gTTS(text=text, lang='en')
tts.save("male_voice.wav")

print("Male voice generated")
```
